```python
# For loop to append each tweet in 'tweet_json.txt' into a list 'list_json_tweets'
tic = time.time()
list_json_tweets = []

tweet_file = open('tweet_json.txt', "r")

for line in tweet_file:
    try:
        tweet = json.loads(line)
        list_json_tweets.append(tweet)
    except:
        continue
        
tweet_file.close()

toc = time.time()
print(toc - tic, ' seconds')
# 0.4938828945159912  seconds
```

```python
#Take a look at first tweet json structure
list_json_tweets[0].keys()
```

```python
# Construct df_info dataframe that includes: tweet id, retweet_count, favorite_count
df_info = pd.DataFrame()
df_info['tweet_id'] = list(map(lambda tweet: tweet['id'], list_json_tweets))
df_info['retweet_count'] = list(map(lambda tweet: tweet['retweet_count'], list_json_tweets))
df_info['favorite_count'] = list(map(lambda tweet: tweet['favorite_count'], list_json_tweets))

# View first few tweets
df_info.head()
```

### Data Assessment Summary
#### 1- Enhanced WeRateDogs Twitter Archive Dataset (df_archive)
The following quality issues were identified:
- Data contains retweets entries mixed with original tweets. Retweets are not of interest.
- Records where 'in_reply_to_status_id' values are not null, are not original tweets and don't have much information.
- The following columns are irrelevant to upcoming analysis: in_reply_to_status_id, in_reply_to_user_id, retweeted_status_id, retweeted_status_user_id and retweeted_status_timestamp
- The timestamp column should be of type datetime instead of object type
- Some tweets have incorrect dog 'stage' information. Manual investigation should be conducted on cases were multiple stages are identified.
- Some tweets contain rating denominator different than 10 and not accurate.
- Some rating numerators have decimals are inaccurate.

The following tidiness issues were identified:
- There is a column for each dog stage (doggo, floofer, pupper and puppo) instead of one cloumn.
- Rating is divided into two columns: numerators and denominators. It can be combined into one as a ratio between numerator and denominator.
- All other datasets should be merged with this dataset

#### 2- The Tweet Image Predictions Dataset (df_predictions)
The following quality issues were identified:
- There are tweets with no images 
- Some tweets have the same images

#### 3- Extra Tweet Information Dataset (df_info)
The following quality issue was identified:
- Some records are related to retweets, which are not of interest

#### Define
Correct inaccurate rating numerator with decimals.
#### Code

```python
#Find rating numerators that have decimals
df_archive_clean[df_archive_clean.text.str.contains(r"(\d+\.\d*\/\d+)")]
```

```python
# Modify rating_numerator and denominator columns to float
df_archive_clean['rating_numerator'] = df_archive_clean['rating_numerator'].astype('float')
df_archive_clean['rating_denominator'] = df_archive_clean['rating_denominator'].astype('float')

#identfy IDs for ratings that have decimal issues
rating_decimal_issues_ids = df_archive_clean.index[df_archive_clean.text.str.contains(r"(\d+\.\d*\/\d+)")].tolist()

#Extract corresponding ratings from tweet's text
ratings = df_archive_clean.loc[rating_decimal_issues_ids].text.str.extract(r"(\d+\.\d*\/\d+)")

#Split ratings to numerator and denominator lists
num, den = list(zip(*(list(map(float, s.split("/"))) for s in ratings)))

#Update ratings
df_archive_clean.loc[rating_decimal_issues_ids,'rating_numerator'] = num
df_archive_clean.loc[rating_decimal_issues_ids,'rating_denominator'] = den
```

From figures below, we see there is a high correlation between retweet and favorite counts. In addition, all tweets have positive retweets and favorites counts, and the account is gaining popularity with time. There some tweets where it was insanely popular.

```python
%matplotlib inline

x=df_master_clean['retweet_count']
y=df_master_clean['favorite_count']

# regression line
fit = np.polyfit(x,y,1)
fit_fn = np.poly1d(fit) 

plt.plot(x, y, "o", x, fit_fn(x), "--k")
plt.title('Retweet vs Favorite Counts')
plt.xlabel('retweet_count')
plt.ylabel('favorite_count')
plt.tight_layout()
plt.savefig('retweet_vs_favorite.png', dpi=200)
```

```python
#convert timestamp to a representative number
ts = (df_master_clean['timestamp'] - np.datetime64('1970-01-01T00:00:00Z')) / np.timedelta64(1, 's')

# create linear regression
fit_retweet = np.polyfit(ts,df_master_clean.retweet_count,1)
fit_favorite = np.polyfit(ts,df_master_clean.favorite_count,1)
fit_retweet_fn = np.poly1d(fit_retweet)
fit_favorite_fn = np.poly1d(fit_favorite)

# store regression lines
df_master_clean['fit_retweet'] = fit_retweet_fn(ts)
df_master_clean['fit_favorite'] = fit_favorite_fn(ts)


fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15,5), sharex=True)

# Plot retweet_count and regression line
df_master_clean.plot(kind='line', color='Blue', x='timestamp', y='retweet_count', ax=axes[0], title='Retweet Count over Time')
df_master_clean.plot(kind='line', color='Red', x='timestamp', y='fit_retweet', ax=axes[0], grid=True)

# Plot favorite_count and regression line
df_master_clean.plot(kind='line', color='Blue', x='timestamp', y='favorite_count', ax=axes[1], title='Favorite Count over Time')
df_master_clean.plot(kind='line', color='Red', x='timestamp', y='fit_favorite', ax=axes[1], grid=True)

#plt.tight_layout()
plt.savefig('retweet_favorite_subplot.png', dpi=200)
plt.show()
```
